<h2 align="left"><font color=#F28888>PROBLEM STATEMENT:</font></h2>

The spread of seasonal flu and H1N1 virus presents significant public health challenges, leading to a need for better understanding of vaccination patterns. Despite the availability of vaccines, some individuals remain unvaccinated, potentially due to various socio-economic, demographic, or behavioral factors. Predicting whether an individual is likely to get vaccinated can help public health organizations target interventions more effectively and improve vaccine uptake, ultimately reducing the spread of infectious diseases.

This project aims to build a predictive model that identifies individuals’ likelihood of receiving the seasonal flu and H1N1 vaccines based on available demographic, health, and behavioral data.

- Build a predictive model to determine whether individuals are likely to receive the seasonal flu and H1N1 vaccines based on key features.

```python
# Checking for null values
def null_value_checker(df):
    missing_values = df.isna().sum().sort_values(ascending=False).to_frame(name='missing_count')

    # Calculating the percentage of missing values
    missing_values['missing_percentage'] = ((missing_values['missing_count'] / len(df)) * 100).round(2)

    return missing_values
    
null_value_checker(new_df)
```

<div style="border-radius:10px; padding: 15px; background-color: rgb(142, 183, 190); font-size:110%; text-align:LEFT; width: 95%; color: BLACK;">

##### <u>Observations and tackling null values in training set</u>
- Health Insurance, employment industry and employment occupation columns have almost half or more than half of their data missing
- The employment industry and employment occupation columns from previous exploration also seems to have incomprehensible values, these 2 will be dropped
- Health insurance which is categorical will be replaced with a 2 for Unspecified
- The rest which are categorical with relatively low missing values will be replaced with the mode or most frequent value.

```python
# Replace double spaces with single spaces in all string columns
def clean_spaces(df):
    # Select object/string columns
    str_cols = df.select_dtypes(include=['object']).columns
    
    # Apply string replacement to each column
    for col in str_cols:
        df[col] = df[col].str.replace(r'\s{2,}', ' ', regex=True)
    
    return df

# Apply to your DataFrame
new_df = clean_spaces(new_df)
```

## <h2><span style='color:#fcc36d; font-size: 15px;'>Step 4.1 |</span><span style='color:#F28888; font-size: 15px;'>Encoding
</span></h2>

```python
# Isolating the object colimns to get the names

cat_cols = new_df.select_dtypes(include = 'object').columns

# viewing the columns

new_df[cat_cols].head()
```

<div style="border-radius:10px; padding: 15px; background-color: rgb(142, 183, 190); font-size:110%; text-align:LEFT; width: 95%; color: BLACK;">



- *education* : Ordinal relationships characteristics of features. For this Label encoder will be used since features have a clear and meaningful order

- *race, hhs_geo_region, census_msa, age_group, income_poverty* : Frequency of this features is particularly useful hence we will use Frequency encoding for this, as for the rest, OHE will be used

```python
# Initialize the label encoder
label_encoder = LabelEncoder()

label_columns = ['education']

for col in label_columns:
    new_df[col] = label_encoder.fit_transform(new_df[col])
```

```python
joblib.dump(label_encoder, f'edu_label_encoder.pkl')
```

```python
# Frequency Encoding with Unique Encoders

# 1. First create encoding dictionaries for each column
race_encoder = new_df['race'].value_counts().to_dict()
age_group_encoder = new_df['age_group'].value_counts().to_dict()
geo_region_encoder = new_df['hhs_geo_region'].value_counts().to_dict()
census_msa_encoder = new_df['census_msa'].value_counts().to_dict()
income_poverty_encoder = new_df['income_poverty'].value_counts().to_dict()

# 2. Apply each encoding separately
new_df['encoded_race'] = new_df['race'].map(race_encoder)
new_df['encoded_age_group'] = new_df['age_group'].map(age_group_encoder)
new_df['encoded_geo_region'] = new_df['hhs_geo_region'].map(geo_region_encoder)
new_df['encoded_census_msa'] = new_df['census_msa'].map(census_msa_encoder)
new_df['encoded_income_poverty'] = new_df['income_poverty'].map(income_poverty_encoder)

# 3. Drop original columns (optional)
new_df.drop(['race', 'age_group', 'hhs_geo_region', 'census_msa', 'income_poverty'], 
            axis=1, inplace=True)

# Verification
print("Encoding completed. New columns:")
print([col for col in new_df.columns if col.startswith('encoded_')])
```

```python
# Dump each encoder to a .pkl file
joblib.dump(race_encoder, 'race_encoder.pkl')
joblib.dump(age_group_encoder, 'age_group_encoder.pkl')
joblib.dump(geo_region_encoder, 'geo_region_encoder.pkl')
joblib.dump(census_msa_encoder, 'census_msa_encoder.pkl')
joblib.dump(income_poverty_encoder, 'income_poverty_encoder.pkl')
```

- Taking the opinions further, I will break them into good and low opinions, we will determine this using the median


```python
def good_opinion(column1,new_column_name, new_df):
    # Calculate the mean of the 'opinion_h1n1' column
    saved_mean = new_df[column1].mean()

    # Create a new column 'h1n1_good_opinion' based on the condition
    new_df[new_column_name] = np.where(new_df[column1] >= saved_mean, 1, 0)

    new_df.drop(columns=column1, axis = 1, inplace= True )
```

```python
# running the function 

good_opinion('opinion_seasonal','Opinion_Scale_seasonal', new_df)
good_opinion('opinion_h1n1','Opinion_Scale_h1n1', new_df)
```

* Good opinion is represented by 1
* Low opinion is represented by 0

```python
# Training and fitting function
def fit_train_score(model, model_filename="trained_model.joblib"):
    # Fit the model
    model.fit(X_train, y_train)

    # Predict with X_test
    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)

    # Calculate accuracy scores
    accuracy_train = accuracy_score(y_train, y_train_pred)
    accuracy_test = accuracy_score(y_test, y_test_pred)

    # Perform cross-validation
    cross_val = cross_val_score(model, X_train, y_train, cv=7)

    # Print the scores
    print(f'Model Train accuracy: {accuracy_train:.3f}')
    print(f'Model Test accuracy: {accuracy_test:.3f}')
    print(f"Cross Validation Accuracy: {cross_val.mean() * 100:.2f}%")

    # Save the model to a file
    joblib.dump(model, model_filename)
    print(f"Model saved as {model_filename}")
    
    print('\n')
    print('***************', '\n')
    print(classification_report(y_test, y_test_pred))
    # Confusion matrix
    cmatrix = confusion_matrix(y_test_pred, y_test)

    plt.figure(figsize=(8, 5))
    ax = sns.heatmap(cmatrix, annot=True, fmt='d', cmap='YlGnBu', cbar=True, linewidths=1, linecolor='black')
    ax.set_title('Confusion Matrix', fontsize=16)
    ax.set_xlabel('Predicted', fontsize=12)
    ax.set_ylabel('True', fontsize=12)

    # Customize tick labels
    ax.xaxis.set_ticklabels(['Class 0', 'Class 1'], fontsize=10)
    ax.yaxis.set_ticklabels(['Class 0', 'Class 1'], fontsize=10)

    plt.tight_layout()
    plt.show()


```

```python

def test_df_transformation(test_df):
    # Make a copy to avoid modifying original
    
    # Select object/string columns
    str_cols = test_df.select_dtypes(include=['object']).columns
    
    # Apply string replacement to each column
    for col in str_cols:
        test_df[col] = test_df[col].str.replace(r'\s{2,}', ' ', regex=True)
    
       
    # 1. Initial column drops
    test_df.drop(['employment_industry', 'employment_occupation'], axis=1, inplace=True)
    
    # 2. Handle missing values
    test_df['health_insurance'].fillna(value=2, inplace=True)
    test_df[:] = imputer.transform(test_df)
    
    # 3. Label encoding
    test_df['education'] = label_encoder.transform(test_df['education'])
    
    # 4. Frequency encoding
    race_encoder = test_df['race'].value_counts().to_dict()
    age_group_encoder = test_df['age_group'].value_counts().to_dict()
    geo_region_encoder = test_df['hhs_geo_region'].value_counts().to_dict()
    census_msa_encoder = test_df['census_msa'].value_counts().to_dict()
    income_poverty_encoder = test_df['income_poverty'].value_counts().to_dict()

    # Apply each encoding separately
    test_df['encoded_race'] = test_df['race'].map(race_encoder)
    test_df['encoded_age_group'] = test_df['age_group'].map(age_group_encoder)
    test_df['encoded_geo_region'] = test_df['hhs_geo_region'].map(geo_region_encoder)
    test_df['encoded_census_msa'] = test_df['census_msa'].map(census_msa_encoder)
    test_df['encoded_income_poverty'] = test_df['income_poverty'].map(income_poverty_encoder)

    # Drop original columns (optional)
    test_df.drop(['race', 'age_group', 'hhs_geo_region', 'census_msa', 'income_poverty'], 
                axis=1, inplace=True)
    
    # One-hot encoding
    columns_to_encode = ['sex', 'marital_status', 'rent_or_own', 'employment_status']
    encoded_data = ohe.transform(test_df[columns_to_encode])
    encoded_df = pd.DataFrame(encoded_data, 
                           columns=ohe.get_feature_names_out(columns_to_encode),
                           index=test_df.index)
    
    test_df = pd.concat([test_df.drop(columns=columns_to_encode), encoded_df], axis=1)
    
    # Behavioral features
# Create the new column 'behavioral_features'
    test_df['behavioral_features'] = (
        test_df['behavioral_antiviral_meds'] + test_df['behavioral_avoidance'] +
        test_df['behavioral_face_mask'] + test_df['behavioral_wash_hands'] +
        test_df['behavioral_large_gatherings'] + test_df['behavioral_outside_home'] +
        test_df['behavioral_touch_face']
    )

    # Drop the old behavioral columns
    columns_to_drop = [
        'behavioral_antiviral_meds', 'behavioral_avoidance', 'behavioral_face_mask', 
        'behavioral_wash_hands', 'behavioral_large_gatherings', 'behavioral_outside_home',
        'behavioral_touch_face'
    ]
    test_df = test_df.drop(columns=columns_to_drop)
    
    # 7. Opinion features
    h1n1_opinion_cols = ['opinion_h1n1_vacc_effective', 'opinion_h1n1_risk', 'opinion_h1n1_sick_from_vacc']
    seasonal_opinion_cols = ['opinion_seas_vacc_effective', 'opinion_seas_risk', 'opinion_seas_sick_from_vacc']
    
    test_df['opinion_h1n1'] = test_df[h1n1_opinion_cols].sum(axis=1)
    test_df['opinion_seasonal'] = test_df[seasonal_opinion_cols].sum(axis=1)
    test_df.drop(columns=h1n1_opinion_cols+seasonal_opinion_cols, inplace=True)
    
    # 8. Opinion scales
    opinion_seasonal_mean = test_df['opinion_seasonal'].mean()
    opinion_h1n1_mean = test_df['opinion_h1n1'].mean()
    test_df['Opinion_Scale_seasonal'] = np.where(test_df['opinion_seasonal'] >= opinion_seasonal_mean, 1, 0)
    test_df['Opinion_Scale_h1n1'] = np.where(test_df['opinion_h1n1'] >= opinion_h1n1_mean, 1, 0)
    test_df.drop(columns=['opinion_seasonal', 'opinion_h1n1'], inplace=True)
    
    # 9. Household features
    test_df['Household_count'] = test_df['household_children'] + test_df['household_adults']
    test_df['Household_count_Scale'] = np.where(test_df['Household_count'] >= 2, 1, 0)
    test_df.drop(columns=['household_children', 'household_adults', 'Household_count'], inplace=True)
    
    # 10. Binary features
    test_df['high_h1n1_concern'] = np.where(test_df['h1n1_concern'] >= 2, 1, 0)
    test_df['high_h1n1_knowledge'] = np.where(test_df['h1n1_knowledge'] >= 1, 1, 0)
    test_df.drop(columns=['h1n1_concern', 'h1n1_knowledge', 'respondent_id'], inplace=True)
    
     
    for col in test_df.columns:
        if col != 'respondent_id':  # Skip respondent_id
            scaler = joblib.load(f'{col}_scaler.pkl')
            test_df[col] = scaler.transform(test_df[[col]]).flatten()
    
    # 12. Select final features
    final_features = [
        'encoded_geo_region', 'behavioral_features', 'encoded_age_group',
        'doctor_recc_seasonal', 'education', 'encoded_census_msa',
        'doctor_recc_h1n1', 'health_insurance'
    ]
    
    return test_df[final_features]
```

```python
%%writefile vaccine_prediction_app.py
import streamlit as st
import joblib
import pandas as pd
import numpy as np
import plotly.express as px
from io import BytesIO
import base64
import os
import traceback

hhs_geo_region = st.selectbox(
    "Geographic Region",
    options=list(geo_region_encoder.classes_)
)

@st.cache_resource
def load_models():
    BASE_DIR = os.path.dirname(os.path.abspath(__file__))

    files = {
        "seasonal_model": "Gradient_boosting_Seasonal_grid.joblib",
        "h1n1_model": "model_xgb_H1N1.joblib",
        "geo_region_encoder": "geo_region_encoder.pkl",
        "age_group_encoder": "age_group_encoder.pkl",
        "edu_label_encoder": "edu_label_encoder.pkl",
        "census_msa_encoder": "census_msa_encoder.pkl",
        "behavioral_features_scaler": "behavioral_features_scaler.pkl",
        "doctor_recc_seasonal_scaler": "doctor_recc_seasonal_scaler.pkl",
        "doctor_recc_h1n1_scaler": "doctor_recc_h1n1_scaler.pkl",
        "health_insurance_scaler": "health_insurance_scaler.pkl"
    }

    loaded_objects = {}

    for name, filename in files.items():
        file_path = os.path.join(BASE_DIR, filename)

        if not os.path.exists(file_path):
            raise FileNotFoundError(f"{filename} not found at {file_path}")

        loaded_objects[name] = joblib.load(file_path)

    return loaded_objects


try:
    models = load_models()
    st.success("All models loaded successfully")

except Exception as e:
    st.error(f"Error loading models: {repr(e)}")
    st.text(traceback.format_exc())


# App config
st.set_page_config(
    page_title="Vaccine Prediction Dashboard",
    page_icon="💉",
    layout="wide"
)

# Set background image to cover entire page
st.markdown(
    """
    <style>
    .stApp {
        background-image: url("https://img.freepik.com/free-vector/coronavirus-globe-concept_52683-35722.jpg?t=st=1744156153~exp=1744159753~hmac=2f129d537b8f4320b204be88ce714a8c8cd3942065f751d1ae02f391cd988702&w=996");
        background-size: cover;
        background-position: center;
        background-repeat: no-repeat;
        background-attachment: fixed;
    }
    /* Sidebar styling */
    section[data-testid="stSidebar"] {
        background-color: rgba(44, 69, 80, 0.85) !important;
        color: white !important;
        backdrop-filter: blur(10px);
    }
    
    /* Sidebar header and text */
    section[data-testid="stSidebar"] h1, 
    section[data-testid="stSidebar"] h2, 
    section[data-testid="stSidebar"] h3, 
    section[data-testid="stSidebar"] h4, 
    section[data-testid="stSidebar"] h5, 
    section[data-testid="stSidebar"] h6, 
    section[data-testid="stSidebar"] p {
        color: white !important;
    }
    
    /* Main content container */
    .main-content {
        background-color: rgba(44, 69, 80, 0.85);
        backdrop-filter: blur(8px);
        border-radius: 15px;
        padding: 2.5rem;
        margin: 2rem;
        box-shadow: 0 8px 32px 0 rgba(31, 38, 135, 0.15);
        border: 1px solid rgba(44, 69, 80, 0.85);
    }
    
    /* Headers */
    h1, h2, h3 {
        color: #FFFFFF;
    }
    
    /* Form elements */
    .stSelectbox, .stRadio, .stCheckbox {
        background-color: #808080;
        border-radius: 10px;
        padding: 12px;
        margin-bottom: 15px;
    }
    
    /* Decision boxes */
    .decision-box {
        border-radius: 10px;
        padding: 15px;
        margin: 15px 0;
        text-align: center;
        font-weight: bold;
        font-size: 1.1rem;
        box-shadow: 0 4px 6px rgba(0,0,0,0.1);
    }
    .likely {
        background: linear-gradient(135deg, rgba(76, 35, 80, 0.9), rgba(129, 199, 132, 0.9));
        color: white;
    }
    .unlikely {
        background: linear-gradient(135deg, rgba(244, 67, 54, 0.9), rgba(229, 115, 115, 0.9));
        color: white;
    }
    .maybe {
        background: linear-gradient(135deg, rgba(255, 193, 7, 0.9), rgba(255, 213, 79, 0.9));
        color: #2c3e50;
    }
    
    /* Submit button */
    .stButton>button {
        background: linear-gradient(135deg, #10b981, #059669);
        color: white;
        font-weight: bold;
        border-radius: 10px;
        padding: 12px 24px;
        border: none;
        transition: all 0.3s ease;
    }
    .stButton>button:hover {
        background: linear-gradient(135deg, #059669, #047857);
        transform: translateY(-2px);
        box-shadow: 0 4px 8px rgba(0,0,0,0.15);
    }

    }
    
    /* Sidebar */
    .sidebar .sidebar-content {
        background-color: rgba(44, 69, 80, 0.85);
        color: white;
        backdrop-filter: blur(10px);
    }
    </style>
    """,
    unsafe_allow_html=True
)

# Create donut chart function
def create_donut_chart(prob, title, color):
    remaining = 1 - prob
    df = pd.DataFrame({
        'Category': ['Vaccinate', 'Not Vaccinate'],
        'Value': [prob, remaining]
    })
    
    fig = px.pie(df, values='Value', names='Category', 
                 hole=0.6, title=title,
                 color_discrete_sequence=[color, 'rgba(240, 240, 240, 0.7)'])
    
    fig.update_traces(textposition='inside', 
                     textinfo='percent+label',
                     hovertemplate=None,
                     marker=dict(line=dict(color='white', width=2)),
                     insidetextfont=dict(color='white'))
    
    fig.update_layout(
        showlegend=False,
        annotations=[dict(text=f"{prob:.0%}", 
                        x=0.5, y=0.5, 
                        font_size=28, 
                        showarrow=False)],
        title_font_color='white',
        font=dict(color='white'),
        margin=dict(t=40, b=0, l=0, r=0),
        paper_bgcolor='rgba(44, 69, 80, 0.85)',
        plot_bgcolor='rgba(200, 200, 200, 0.85)'
    )
    
    return fig

# Function to create download link
def get_table_download_link(df):
    csv = df.to_csv(index=False)
    b64 = base64.b64encode(csv.encode()).decode()
    href = f'<a href="data:file/csv;base64,{b64}" download="vaccine_predictions.csv">Download CSV with predictions</a>'
    return href

# Main content container
st.markdown(
    """
    <style>
        /* Main background */
        .stApp {
            background-color: #000000;
        }
        
        /* All text elements */
        h1, h2, h3, h4, h5, h6, p, .stMarkdown, .stText, .stNumberInput label, 
        .stSelectbox label, .stSlider label, .stRadio label {
            color: black !important;
        }
        
        /* Specific header styling */
        h1 {
            font-size: 2.5rem;
            margin-bottom: 1rem;
        }
        
        /* Remove extra whitespace */
        .block-container {
            padding-top: 1rem;
            padding-bottom: 1rem;
        }
    </style>
    """,
    unsafe_allow_html=True
)

# App header
st.title("Vaccine Uptake Prediction")
st.markdown("Predict the likelihood of receiving H1N1 and seasonal flu vaccines")

# Create tabs
tab1, tab2 = st.tabs(["Single Prediction", "Batch Prediction"])

with tab1:
    # Create form
    with st.form("vaccine_form"):
        st.subheader("Personal Information")
        
        col1, col2 = st.columns(2)
        
        with col1:
            hhs_geo_region = st.selectbox(
                "Geographic Region",
                options=list(geo_region_encoder.keys())
            )
            
            age_group = st.selectbox(
                "Age Group",
                options=list(age_group_encoder.keys())
            )
            
            census_msa = st.selectbox(
                "Metropolitan Status",
                options=list(census_msa_encoder.keys())
            )
        
        with col2:
            education = st.selectbox(
                "Education Level",
                options=list(edu_label_encoder.classes_)
            )
            
            health_insurance = st.radio(
                "Has Health Insurance",
                options=[1, 0],
                format_func=lambda x: "Yes" if x else "No",
                horizontal=True
            )
        
        st.subheader("Doctor Recommendations")
        col1, col2 = st.columns(2)
        with col1:
            doctor_recc_seasonal = st.radio(
                "Recommended Seasonal Vaccine",
                options=[1, 0],
                format_func=lambda x: "Yes" if x else "No",
                horizontal=True
            )
        with col2:
            doctor_recc_h1n1 = st.radio(
                "Recommended H1N1 Vaccine",
                options=[1, 0],
                format_func=lambda x: "Yes" if x else "No",
                horizontal=True
            )
        
        st.subheader("Protective Behaviors")
        st.markdown("Select all behaviors practiced in the last month:")
        
        cols = st.columns(2)
        with cols[0]:
            behavioral_antiviral_meds = st.checkbox("Took antiviral medications")
            behavioral_avoidance = st.checkbox("Avoided crowded places")
            behavioral_face_mask = st.checkbox("Wore face mask")
        with cols[1]:
            behavioral_wash_hands = st.checkbox("Frequently washed hands")
            behavioral_large_gatherings = st.checkbox("Avoided large gatherings")
            behavioral_outside_home = st.checkbox("Reduced time outside home")
        behavioral_touch_face = st.checkbox("Avoided touching face")
        
        submitted = st.form_submit_button("Predict Vaccine Uptake")

    if submitted:
        try:
            # Calculate behavioral features sum
            behavioral_features = sum([
                int(behavioral_antiviral_meds),
                int(behavioral_avoidance),
                int(behavioral_face_mask),
                int(behavioral_wash_hands),
                int(behavioral_large_gatherings),
                int(behavioral_outside_home),
                int(behavioral_touch_face)
            ])
            
            # Apply individual scaling
            behavioral_features_scaled = behavioral_scaler.transform([[behavioral_features]])[0][0]
            doctor_recc_seasonal_scaled = seasonal_rec_scaler.transform([[doctor_recc_seasonal]])[0][0]
            doctor_recc_h1n1_scaled = h1n1_rec_scaler.transform([[doctor_recc_h1n1]])[0][0]
            health_insurance_scaled = health_insurance_scaler.transform([[health_insurance]])[0][0]
            
            # Create input dataframe
            input_data = pd.DataFrame({
                'encoded_geo_region': [geo_region_encoder[hhs_geo_region]],
                'behavioral_features': [behavioral_features_scaled],
                'encoded_age_group': [age_group_encoder[age_group]],
                'education': [edu_label_encoder.transform([education])[0]],
                'doctor_recc_seasonal': [doctor_recc_seasonal_scaled],
                'doctor_recc_h1n1': [doctor_recc_h1n1_scaled],
                'encoded_census_msa': [census_msa_encoder[census_msa]],
                'health_insurance': [health_insurance_scaled]
            })
            
            # Make predictions
            seasonal_prob = seasonal_model.predict_proba(input_data)[0][1]
            h1n1_prob = h1n1_model.predict_proba(input_data)[0][1]
            
            # Determine decisions
            def get_decision(prob):
                if prob >= 0.6:
                    return "LIKELY TO VACCINATE", "likely"
                elif prob >= 0.4:
                    return "MAY VACCINATE", "maybe"
                else:
                    return "UNLIKELY TO VACCINATE", "unlikely"
            
            seasonal_decision, seasonal_class = get_decision(seasonal_prob)
            h1n1_decision, h1n1_class = get_decision(h1n1_prob)
            
            # Display results
            st.subheader("Prediction Results")
            
            col1, col2 = st.columns(2)
            with col1:
                st.markdown("<div class='chart-container'>", unsafe_allow_html=True)
                st.plotly_chart(create_donut_chart(seasonal_prob, "Seasonal Vaccine", "#3498db"), 
                              use_container_width=True)
                st.markdown("</div>", unsafe_allow_html=True)
                st.markdown(
                    f'<div class="decision-box {seasonal_class}">{seasonal_decision}</div>',
                    unsafe_allow_html=True
                )
            
            with col2:
                st.markdown("<div class='chart-container'>", unsafe_allow_html=True)
                st.plotly_chart(create_donut_chart(h1n1_prob, "H1N1 Vaccine", "#e74c3c"), 
                              use_container_width=True)
                st.markdown("</div>", unsafe_allow_html=True)
                st.markdown(
                    f'<div class="decision-box {h1n1_class}">{h1n1_decision}</div>',
                    unsafe_allow_html=True
                )

        except Exception as e:
            st.error(f"An error occurred during prediction: {str(e)}")

with tab2:
    st.header("Batch Predictions from CSV")
    st.markdown("Upload a CSV file with multiple records to get predictions for all entries")
    
    uploaded_file = st.file_uploader("Choose a CSV file", type="csv")
    
    if uploaded_file is not None:
        try:
            # Read the uploaded file
            df = pd.read_csv(uploaded_file)
            df = df.astype(str)
            
            # Store original columns for later
            original_columns = df.columns.tolist()
            
            # Validate columns
            required_columns = [
                'hhs_geo_region', 'age_group', 'census_msa', 'education',
                'health_insurance', 'doctor_recc_seasonal', 'doctor_recc_h1n1'
            ]
            
            behavioral_columns = [
                'behavioral_antiviral_meds', 'behavioral_avoidance',
                'behavioral_face_mask', 'behavioral_wash_hands',
                'behavioral_large_gatherings', 'behavioral_outside_home',
                'behavioral_touch_face'
            ]
            # Convert Yes/No to 1/0 for binary columns
            binary_columns = [
                'health_insurance', 'doctor_recc_seasonal', 'doctor_recc_h1n1',
                *behavioral_columns  # Unpack behavioral columns
            ]
            
            yes_no_map = {
                'yes': '1',
                'no': '0',
                'y': '1',
                'n': '0',
                '1': '1',  # Keep existing 1s
                '0': '0'   # Keep existing 0s
            }
            
            for col in binary_columns:
                if col in df.columns:
                    # Case-insensitive conversion
                    df[col] = df[col].str.lower().map(yes_no_map).fillna(df[col])
                    # Convert to numeric
                    df[col] = pd.to_numeric(df[col], errors='coerce')
                    # Fill any remaining invalid values with 0
                    df[col] = df[col].fillna(0).astype(int)
            
            
            # Now validate for missing values
            if df.isnull().any().any():
                missing_cols = df.columns[df.isnull().any()].tolist()
                st.error(f"Missing values detected in columns: {', '.join(missing_cols)}")
                st.stop()
            
            missing_columns = [col for col in required_columns if col not in df.columns]
            
            if missing_columns:
                st.error(f"Missing required columns: {', '.join(missing_columns)}")
            else:
                # Process the data similar to single prediction
                # Encode categorical variables
                df['encoded_geo_region'] = df['hhs_geo_region'].map(geo_region_encoder)
                df['encoded_age_group'] = df['age_group'].map(age_group_encoder)
                df['encoded_census_msa'] = df['census_msa'].map(census_msa_encoder)
                df['education'] = edu_label_encoder.transform(df['education'])
                
                # Calculate behavioral features sum
                behavioral_cols_present = [col for col in behavioral_columns if col in df.columns]
                if behavioral_cols_present:
                    df['behavioral_features'] = df[behavioral_cols_present].sum(axis=1)
                else:
                    df['behavioral_features'] = 0
                
                # Scale features
                df['behavioral_features'] = behavioral_scaler.transform(df[['behavioral_features']])
                df['doctor_recc_seasonal'] = seasonal_rec_scaler.transform(df[['doctor_recc_seasonal']])
                df['doctor_recc_h1n1'] = h1n1_rec_scaler.transform(df[['doctor_recc_h1n1']])
                df['health_insurance'] = health_insurance_scaler.transform(df[['health_insurance']])
                
                # Prepare input data
                input_data = df[[
                    'encoded_geo_region', 'behavioral_features', 'encoded_age_group', 'education',
                    'doctor_recc_seasonal', 'doctor_recc_h1n1',
                    'encoded_census_msa', 'health_insurance'
                ]]
                
                # Make predictions
                seasonal_probs = seasonal_model.predict_proba(input_data)[:, 1]
                h1n1_probs = h1n1_model.predict_proba(input_data)[:, 1]
                
                # Add predictions to dataframe
                df['seasonal_vaccine_prob'] = seasonal_probs
                df['h1n1_vaccine_prob'] = h1n1_probs
                
                # Add decision categories
                def get_decision_category(prob):
                    if prob >= 0.6:
                        return "Likely"
                    elif prob >= 0.4:
                        return "Maybe"
                    else:
                        return "Unlikely"
                
                df['seasonal_decision'] = df['seasonal_vaccine_prob'].apply(get_decision_category)
                df['h1n1_decision'] = df['h1n1_vaccine_prob'].apply(get_decision_category)
                
                # Show preview of results
                st.subheader("Prediction Results Preview")
                st.dataframe(df[original_columns + ['seasonal_vaccine_prob', 'h1n1_vaccine_prob', 
                                                   'seasonal_decision', 'h1n1_decision']].head())
                
                # Show summary statistics
                st.subheader("Prediction Summary")
                
                col1, col2 = st.columns(2)
                with col1:
                    st.metric("Average Seasonal Vaccine Probability", 
                             f"{df['seasonal_vaccine_prob'].mean():.1%}")
                    st.metric("Likely to Take Seasonal Vaccine", 
                             f"{len(df[df['seasonal_decision'] == 'Likely'])} people")
                    
                    # Seasonal vaccine distribution chart
                    fig1 = px.pie(df, names='seasonal_decision', title='Seasonal Vaccine Decisions',
                                 color_discrete_sequence=px.colors.sequential.Blues_r)
                    st.plotly_chart(fig1, use_container_width=True)
                
                with col2:
                    st.metric("Average H1N1 Vaccine Probability", 
                             f"{df['h1n1_vaccine_prob'].mean():.1%}")
                    st.metric("Likely to Take H1N1 Vaccine", 
                             f"{len(df[df['h1n1_decision'] == 'Likely'])} people")
                    
                    # H1N1 vaccine distribution chart
                    fig2 = px.pie(df, names='h1n1_decision', title='H1N1 Vaccine Decisions',
                                color_discrete_sequence=px.colors.sequential.Reds_r)
                    st.plotly_chart(fig2, use_container_width=True)
                
                # Age group analysis
                st.subheader("Vaccine Probability by Age Group")
                age_group_analysis = df.groupby('age_group')[['seasonal_vaccine_prob', 'h1n1_vaccine_prob']].mean().reset_index()
                fig3 = px.bar(age_group_analysis, x='age_group', y=['seasonal_vaccine_prob', 'h1n1_vaccine_prob'],
                             barmode='group', title='Average Vaccine Probability by Age Group',
                             labels={'value': 'Probability', 'variable': 'Vaccine Type'})
                st.plotly_chart(fig3, use_container_width=True)
                
                # Create download link
                st.markdown(get_table_download_link(df), unsafe_allow_html=True)
                
        except Exception as e:
            st.error(f"An error occurred during batch processing: {str(e)}")

st.markdown("</div>", unsafe_allow_html=True)

# Sidebar information
with st.sidebar:
    st.header("About This Tool")
    st.markdown("""
    This dashboard predicts vaccine uptake probability based on:
    - Demographic factors
    - Doctor recommendations
    - Protective behaviors
    
    **Interpretation Guide:**
    - **≥60%**: Very likely to vaccinate
    - **40-59%**: May vaccinate
    - **<40%**: Unlikely to vaccinate
    """)
    st.markdown("---")
    st.markdown("**Model Information**")
    st.markdown("""
    - Uses Gradient Boosting models
    - Trained on CDC data
    - 8 key predictive factors
    """)
    

# Footer
st.markdown("""
<div style="text-align: center; color: white; margin-top: 20px;">
    © 2025 Vaccine Prediction Dashboard | For public health use
</div>
""", unsafe_allow_html=True)
```
